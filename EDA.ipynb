```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path_train = "./datasets/hospitalizaciones_train.csv"
train_data = pd.read_csv(file_path_train, index_col='patientid')
train_data.head()
```

```python
train_data["Stay"] = (train_data["Stay (in days)"] > 8).astype(int)
train_data.head()
```

```python
plt.figure(figsize=(10,6))
plt.title("titulo")
sns.scatterplot(x=train_data['Admission_Deposit'], y=train_data['Available Extra Rooms in Hospital'])
```

```python
sns.histplot(data=train_data, x='Admission_Deposit')
```

```python
sns.histplot(data=train_data, x='Admission_Deposit', hue='Type of Admission')
```

```python
sns.histplot(data=train_data, x='Admission_Deposit', hue='Severity of Illness')
```
